```python
# --- NYT API KEY ---
NYT_API_KEY = "jM4y6U4wMyGSW6rV0D2e2oGTi00hHlzG"
assert NYT_API_KEY, "NYT_API_KEY is empty"

# Fetch Top Stories → DataFrame
import requests, pandas as pd
from pandas import json_normalize

def fetch_topstories(section="technology", api_key=NYT_API_KEY):
    url = f"https://api.nytimes.com/svc/topstories/v2/{section}.json"
    r = requests.get(url, params={"api-key": api_key}, timeout=30)
    
    if r.status_code != 200:
        try:
            print("Server message:", r.json())
        except Exception:
            print("Raw text:", r.text[:300])
    r.raise_for_status()
    return r.json()

raw = fetch_topstories("technology")
items = raw.get("results", [])

# Normalize JSON → flat table
df = json_normalize(items)
keep = ["section","subsection","title","byline","abstract","url","published_date","updated_date"]
df = df[[c for c in keep if c in df.columns]].copy()
for col in ["published_date","updated_date"]:
    if col in df:
        df[col] = pd.to_datetime(df[col], errors="coerce")

print("Rows:", len(df))
df.head(10)


```
